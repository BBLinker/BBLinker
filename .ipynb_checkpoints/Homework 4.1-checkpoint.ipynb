```python
%matplotlib inline
import numpy as np
import pandas as pd
from matplotlib.font_manager import FontProperties

import matplotlib.pyplot as plt
plt.rcParams['font.sans-serif'] = ['Microsoft JhengHei'] 
plt.rcParams['axes.unicode_minus'] = False
```

```python
import matplotlib as mpl
```

```python
df=pd.read_csv('https://quality.data.gov.tw/dq_download_csv.php?nid=29935&md5_url=01410d8f98b8e14419aab3ad9802e371')
```

# 將資料弄出來

```python
df.head()
```

# 先刪掉後面看不到的部分

```python
df.loc[0:21,'日期':'農林漁牧用電佔比(%)']
```

```python
df2=df.loc[0:21,'日期':'農林漁牧用電佔比(%)']
```

# 用drop丟掉不需要的部分

```python
df2.drop(['住宅部門用電佔比(%)','服務業部門用電佔比(%)','機關用電用電佔比(%)','農林漁牧用電佔比(%)'],axis=1)
```

```python
df3=df2.drop(['住宅部門用電佔比(%)','服務業部門用電佔比(%)','機關用電用電佔比(%)','農林漁牧用電佔比(%)'],axis=1)
```

```python
df3['民生用電量']=df3['住宅部門售電量(度)']
```

```python
df3.head()
```

```python
df3['工業用電量']=df3['服務業部門售電量(度)']+df3['機關用電售電量(度)']+df3['農林漁牧售電量(度)']
```

```python
df3.head()
```

# 插入新資訊 '民生用電量/工業用電量'

```python
df3['民生用電量/工業用電量']=df3['民生用電量']/df3['工業用電量']
```

```python
df3.head()
```

```python
c=df3['縣市']
```

```python
e=np.array(c)
```

```python
e
```

```python
d=df3['民生用電量/工業用電量']
```

# 繪製成圖

```python
plt.figure(figsize=(21,7))
plt.bar(e,d)
plt.title("民生用電量/工業用電量",size=40)
plt.ylabel("比率",size=20) 
plt.xlabel("各縣市",size=20)  
plt.show()
```
