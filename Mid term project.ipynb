```python
%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt
from IPython.display import clear_output
import time
```

利用判斷式來完成老虎機，並且為了美觀使用clear_output來刪掉之前的變數

為了呈現老虎機的感覺，使用time.sleep來減緩輸出時間，並且使用多個矩陣來呈現老虎機那種先輸出第一行再來第二行、第三行的感覺。

增加能夠bet的金額，並且會隨著bet的不同來增加能夠連線的條數
