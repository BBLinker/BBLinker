```python
df.drop(['No','X1 transaction date','X2 house age','X4 number of convenience stores','X5 latitude','X6 longitude'],axis=1)
```

```python
df2=df.drop(['No','X1 transaction date','X2 house age','X4 number of convenience stores','X5 latitude','X6 longitude'],axis=1)
```

```python
from sklearn.model_selection import train_test_split
```

```python
x=np.array(df2['X3 distance to the nearest MRT station'])
```

```python
x
```

```python
x.reshape(len(x),1)
```

```python
X=x.reshape(len(x),1)
```

```python
y=np.array(df2['Y house price of unit area'])
```

```python
y
```

```python
Y=y.reshape(len(y),1)
```

```python
Y
```

```python
x_train,x_test,y_train,y_test=train_test_split(X,Y,test_size=0.3,random_state=9487)
```

```python
x_train
```

```python
plt.scatter(x_train,y_train)
```

```python
regr.fit(x_train,y_train)
```

```python
Y_pred=regr.predict(x_test)
```

```python
plt.figure(figsize=(18,7))
plt.scatter(x_test,y_test)
plt.plot(x_test,Y_pred,'r')
plt.xlabel("The nearest MRT station",size=20)
plt.ylabel("Price per unit area",size=20) 
```
