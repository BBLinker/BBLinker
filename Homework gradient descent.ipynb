```python
%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt
import sympy as sp
```

# 先隨便寫一個目標函數並且畫出圖

```python
x=np.linspace(0,1,100)
def f(x):
    return x**4+4*x**3-2*x+1
y=f(x)
plt.plot(x,y)
```

# 想讓電腦自動幫忙算微分引入sympy

```python
z=sp.symbols('z')
f_1=z**4+4*z**3-2*z+1
f_2=sp.diff(f_1,z)
```

# 定義gradient_desent函數，因為剛剛是用電腦幫忙算微分，但是不知道怎麼寫才能讓電腦帶值(我不會處理symbol)，因此引入sp.limit因為連續函數的極限值等於函數值

```python
def gradient_descent(x_start,epoch,learning_rate):
    A=[]
    y_A=[]
    x=x_start
    for i in range (epoch):
        x=x-learning_rate*sp.limit(f_2,z,x)
        A.append(x)
        y_A.append(f(x))
        plt.scatter(A[i], y_A[i], c='r')
```

# 從1開始逼近

```python
h=gradient_descent(1,100,0.01)
plt.plot(x,y)
```

# 從0開始

```python
h=gradient_descent(0,100,0.01)
plt.plot(x,y)
```
